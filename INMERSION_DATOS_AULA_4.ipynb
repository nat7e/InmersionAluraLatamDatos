```python
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Area']
inmuebles.head(3)
```

```python
inmuebles.groupby('Barrio').mean()
```

```python
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
# Pasar los valores calculados de datos_barrio a inmuebles. Como diccionario.
# m2_barrio tiene una salida :  Ejemplo - 'Castilla:552211'
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']

# map(m2_barrio): toma todos los valores del diccionario que se tomo 'm2_barrio' 
# y los va a colocar como traduccion. El map hace el papale de enlar ya que 'Valor_m2_barrio'
# contiene el nombre de los barrio y no el valor usando map(m2_barrio) donde m2_barrio contiene
# informacion como un diccionario, al usar inmuebles['Valor_m2_Barrio'].map(m2_barrio) como el
# 'Valor_m2_Barrio' contiene nombre de barrio y m2_barrio el nombre de barrio y el numero en m2. 
# da una salida de solo el numero en m2 como se oberva.
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
# Al usar [:10] en esta serie, se seleccionan los 10 valores mas comunes en la columna Barrio. 
# index es para obtener solamente los nombres y si quieres los valores usar 'values'. Se hace esto para 
# estudiar los 10 mas nombrados en el barrio.
top_barrios = inmuebles['Barrio'].value_counts()[:10].index

# Al usar reset_index elimina el indice actual de un DataFrame y lo reemplaza
# por un indice numerico predeterminado. Esto crea una nueva columna llamada index que
# contiene el indice anterior.
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
# Esta linea filtra las filas de datos_barrio que tienen un valor en la coliumna Barrio que coincide
# con uno de los valores de la lista top_barrios.
datos_barrio.query('Barrio in @top_barrios')
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x='Barrio', y='Valor_m2_Barrio', data= datos_barrio.query('Barrio in @top_barrios'))

# Para modificar las marcas y etiquetas del eje x de un grafico.
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios'))

# Para modificar las marcas y etiquetas del eje x de un grafico. Una manera de interpretar la grafica es
# en el rectangulo la linea inferior representa el 25%, la linea del medio representa la mitad y la tercera
# representa el 75%. Los puntos que estan mas arriba, indica que son valores que estan fuera de lo comun 
# como por ejemplo el numero de habitaciones de 100 en una casa. se conoce como outlayers.
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
# En el codigo al agregar '& Valor_m2_Millon < 15' hace que los valores no pasen de 15 millones de valor_m2_Millon
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios & Valor_m2_Millon < 15'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Ahora hacemos con el Area
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Area', data = inmuebles.query('Barrio in @top_barrios  & Area < 500'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Se importa Scikitlearn
from sklearn.linear_model import LinearRegression

# importa que permite hacer una separacion de nuestro conjuento de datos en datos para entrenamiento
# y datos para prueba. No solo se puede modelar sino tambien validar.
from sklearn.model_selection import train_test_split

# Una buena base para realizar es tener una linea base
X = datos_ml[['COD_UPZ_GRUPO']]
y = datos_ml['Precio_Millon']

# Debemos tener X de entrenamiento X de pruebas ademas y de entrenamiento e y de pruebas y se hacet el
# split con train_test_split
# test_size : el tamaño se set que deseas. pode defecto es el 25%. Que entrena con el 75% ye l 25% prueba
# random_state=99: Estado de aleatoridad para garantizar la reproducitibilidad de los datos. Este es un modelo
# probabilistico. Actua como una semilla.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=99)
```

```python
# Se utiliza para implementar diferentes algoritmos de aprendizaje automatico.

# Se puede usar esta instancia para entrenar un modelo de regresion lineal en nuestro datos de
# entrenamiento y hacer predicciones en los datos de prueba.

# Aqui una variable la cual se instancia la regresion lineas de forma generica.
modelo = LinearRegression()
```

```python
# fit ajusta  el modelo a los datos de entrenamiento, es decir para encontrar la relacion entre las
# caracteristicas o variables independientes X_train.

# El metodo fit() ajustara el modelo a los datos de entrenamiento X_train y y_train encontrando los 
# coeficientes optimos para los parametros del modelo.

# Una vez que el modelo ha sido entrenado, se puede utilizar para hacer predicciones sobre datos nuevos
# o no vistos.
modelo.fit(X_train,y_train)
```

```python
# Es utilizado para hacer predicciones en el conjunto de datos de prueba (X_test). El modelo  entrenado
# de la etapa anterior (modelo.fit(X_train, y_train)) seutiliza para hacer estas predicciones.

# La variable y_predict contiene las predicciones del modelo para el conjunto de datos de prueba('X_test')
# estas predicciones son hechas con el fin de evaluar el rendimiento del modelo y determinar que tan bien 
# puede generalizar a datos nuevo y no vistos previamente.
y_predict_test = modelo.predict(X_test)
```

```python
# Etapa de validacion.
# error absoluto medio: toma la diferencia de valores que se puede encontrar al hacer el calculo.
# r2 el coeficiente de determinacion que permite saber que tan bien va nuestro modelo, va de 0 1. siendo 1 mejor.

from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# Se instancia el valor absoluto medio. Compara los datos que el modelo nunca vio con lo que se 
# predijo con la ml(machine learning)
baseline_mae = mean_absolute_error(y_test, y_predict_test)
baseline_mae
# Esto nos dice que si introducimos el valor de un inmueble el va a tener una variacion de 349 millones
# por lo que esto esta fuera de nuestro contexto.
```

```python
baseline_r2 = r2_score(y_test, y_predict_test)
baseline_r2
# Esta salida tan baja quiere decir que nuestro modelo no se esta desempeñando nada bien
```
